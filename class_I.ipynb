# The Time Value of Money: Interest and Annuities

This lesson prompts students to solve basic problems about the time value of money.  These problems are meant to begin familiarizing you with the problem solving process, communicating your results, using technology like google sheets, and formalizing your intuition using mathematical formulas.

**OBJECTIVES**

- Solve problems involving simple interest
- Solve problems involving periodic compounding
- Solve problems with ordinary annuities

## Problems: Simple Interest

For each of the problems below, determine the solution and explain your reasoning.

1. Suppose you receive \$500.00 and invest it for 24 months in an account that has an 8% interest rate applied each year.

   - How much money is in the account after the two years?  Explain your reasoning.
  
2. Revisit the first problem, this time the accounts interest is compounded quarterly.

   - How do you compute the interest rate for each quarter?
   - What is the amount in the account at the end of two years?

3. Which of the following options would you prefer?

- Option A: Receive \$98.00 today
- Option B: Receive \$106.33 one year from today
